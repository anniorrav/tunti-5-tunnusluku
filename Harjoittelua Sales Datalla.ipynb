```python
df=pd.read_csv('http://taanila.fi/sales_data_sample.csv', encoding ='cp1252') 
df.head()
```

```python
CITY = ['NYC', 'Reims', 'Paris', 'Pasedena', 'San Francisco']
COUNTRY = ['USA', 'France']
DEALSIZE = ['Small', 'Medium']
```

```python
df7 = df.pivot_table(values = 'SALES', index = ['COUNTRY', 'CITY'], columns = 'DEALSIZE')
df7.style.format('{:.0f}')
```

```python
df7 = df.pivot_table(values = 'SALES', index = ['COUNTRY'], columns = 'DEALSIZE')
df7.style.format('{:.0f}')
```

```python
df7.plot.barh()
plt.xlabel('SALES')
```

```python
ax = df.boxplot('SALES', by = 'COUNTRY')
plt.title('')
plt.suptitle('')
plt.xlabel('COUNTRY')
plt.ylabel('SALES')
ax.set_xticklabels(COUNTRY)

```
